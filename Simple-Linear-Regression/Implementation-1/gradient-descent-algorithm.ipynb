# Gradient Descent

```python
def function(b): # y = (b-4)^2 # change b -> y min
    return (b-4)**2
```

```python
def derivative(b):
    return 2*(b-4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2,10,100)
y = function(x)
plt.plot(x,y)
plt.show()
```

```python
b = 9
# Gradient Descent: 
for i in range(100): # number of iteration
    # dfunction/db:
    b = b - .1*derivative(b) # learning_rate- small
    print(b)
```
